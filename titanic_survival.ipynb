## Titanic Survival Prediction Using Logistic Regression

#### STEPS:
1. Load Data
2. Fix the NULL values (Drop and Replace)
3. Replace(Map) Strings values to Integer
4. Drop the Unnecessary Columns from the dataset

--> Perform the same for both train and test data...calculate the score

--> Predict Y_test for the given X_test

```python
import pandas as pd
```

```python
train_data=pd.read_csv("E:\ML BASIC\logistic_project/train.csv")
test_data=pd.read_csv("E:\ML BASIC\logistic_project/test.csv")
```

```python
# train_data.head()    
test_data.head()
```

```python
train_data.isnull().sum()
```

```python
#drop Cabin column as it has lots of null values
train_data=train_data.drop("Cabin",axis=1)
```

```python
train_data.isnull().sum()
```

```python
#fill null value of age with its mean
train_data['Age'].fillna(train_data['Age'].mean(),inplace=True)
```

```python
#fill Embarked Null with the most appeared value
train_data['Embarked'].fillna(train_data['Embarked'].mode()[0],inplace=True)
```

```python
#check if any null present 
train_data.isnull().sum()
```

```python
#convert string values in Sex and Embarked to integer
train_data.replace({'Sex':{'male':0,'female':1},'Embarked':{'S':0,'C':1,'Q':2}},inplace=True)
```

```python
train_data.head()
```

```python
X_train=train_data.drop(columns=['Name','Ticket','Survived'],axis=1)
```

```python
X_train
```

```python
Y_train=train_data['Survived']
```

```python
Y_train
```

```python
test_data
```

## We need perform the same operation to get x_test

```python
test_data.isnull().sum()
```

```python
#drop cabin
test_data=test_data.drop('Cabin',axis=1)

#replace Nan age by mean
test_data['Age'].fillna(test_data['Age'].mean(),inplace=True)

#replace Nan Embarked by most appeared
test_data['Embarked'].fillna(test_data['Embarked'].mode()[0],inplace=True)

#replace sex and embarked value by integer
test_data.replace({'Sex':{'male':0,'female':1},'Embarked':{'S':0,'C':1,'Q':2}},inplace=True)

#drop name ticket
test_data.drop(columns=['Name','Ticket'],inplace=True)
```

```python
X_test=test_data
```

```python
X_test
```

```python
X_train
```

```python
Y_train
```

```python
#make object of logistic Regression
from sklearn.linear_model import LogisticRegression
model=LogisticRegression()
```

```python
model.fit(X_train,Y_train)
```

```python
model.score(X_train,Y_train)
```

```python
Y_pred=model.predict(X_test)
```
